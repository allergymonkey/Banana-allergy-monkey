```python
import pandas as pd
df = pd.read_csv ('cutebohyun.csv', index_col=0, encoding = 'cp949')
df.T.to_csv('cutebohyun.csv',encoding = 'cp949')
import csv
data = csv.reader(open('cutebohyun.csv'))
next(data)
data = list(data)

people=[]


city = input ("당신이 살고있는 도시는 어디죠? (시 별로 입력해주세요 Ex 서울) : ")


for row in data :
    if city in row :
        for i in range(1,len(row)) :
            people.append(row[i])
            
           
import matplotlib.pyplot as plt
plt.figure(figsize=(6,4), dpi=300)

labels = '살인','살인미수','강도','강간','강제추행','방화','절도','상해','폭행','협박','폭력행위','공갈','손괴','사기','횡령','교통범죄'
explode = (1.5,0,0.8,0.5,0,0.6,0,0,0,0,0,0.9,0,0,0,0)

SMALL_SIZE = 4

plt.title('당신의 도시의 범죄율 원 그래프')
plt.rc('font',family='Malgun Gothic',size=SMALL_SIZE)
plt.pie(people,labels=labels, explode=explode,autopct='%0.1f%%')
plt.axis('equal')
plt.show()
```
